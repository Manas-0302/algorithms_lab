# 1. Searching
Implement Linear Search. Determine the time required to search for element. Repeat the experiment for different values of n, the number of elements in the list to besearched and plot a graph of the time taken versus n.

- To implement Linear Search, measure its time complexity for searching an element, vary the input size by changing the number of elements in the list, and plot a graph to visualize the relationship between time taken and input size
- Algorithm:  
1. Start with the first element of the list.
2. Compare the element with the value being searched for.
3. If the element matches, return the index.
4. If the element does not match, move to the next element and repeat steps 2-3.
5. If the end of the list is reached and no match is found, return -1

- **Created by Dr. Ajay

```python
# write code for Linear Search and plot a graph to visualize the relationship between time taken and input size
import time
import random
import matplotlib.pyplot as plt


def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1


input_sizes = [1000, 2000, 5000, 10000, 20000, 50000]
time_taken = []

for n in input_sizes:
    arr = [random.randint(1, n) for _ in range(n)]
    key = -1

    start = time.time()
    linear_search(arr, key)
    end = time.time()

    time_taken.append(end - start)


plt.plot(input_sizes, time_taken, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken (seconds)")
plt.title("Linear Search Time Complexity (O(n))")
plt.grid(True)
plt.show()
```

# 2. Searching
Implement recursive Binary Search. Determine the time required to search an element.Repeat the experiment for different values of n, the number of elements in the list to besearched and plot a graph of the time taken versus n.
- implement recursive Binary Search, measure its time complexity for searching an element and plot a graph of time taken versus different values of n, the number of elements in the list.  
Algorithm:
1. Start with the middle element of the list
2. If the middle element matches the element being searched, return the index of the middle element
3. If the middle element is greater than the element being searched, recursively call the function on the left half of the list
4. If the middle element  is  less than the element  being searched, recursively call the function on the right half of the list
5. If the element is not found, return -1

```python
#Prob_2: write code for recursive Binary Search and plot a graph to visualize the relationship between time taken and input size
import time
import random
import matplotlib.pyplot as plt


def binary_search(arr, low, high, key):
    if low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] > key:
            return binary_search(arr, low, mid - 1, key)
        else:
            return binary_search(arr, mid + 1, high, key)
    return -1


input_sizes = [1000, 2000, 5000, 10000, 20000, 50000]
time_taken = []


for n in input_sizes:
    arr = list(range(n))
    key = -1

    start = time.time()
    binary_search(arr, 0, n - 1, key)
    end = time.time()

    time_taken.append(end - start)


plt.plot(input_sizes, time_taken, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken (seconds)")
plt.title("Recursive Binary Search Time Complexity (O(log n))")
plt.grid(True)
plt.show()
```

#Prob-3
Given a text txt [0...n-1] and a pattern pat [0...m-1], write a function search (char pat [ ],char txt [ ]) that prints all occurrences of pat [ ] in txt [ ]. You may assume that n > m.
- Aim:  To write a function "search" that takes a pattern and a text as input, and outputs all occurrences of the pattern in the text.  Algorithm:
1. Initialize an empty list to store the indices where the pattern occurs in the text.
2. Loop through each character in the text. For each character, check if the substring starting at that character matches the pattern.
3. If there is a match, append the starting index of the substring to the list of occurrences.
4. Continue looping through the text until all possible substrings have been checked.
5. Return the list of indices where the pattern occurs in the text.  




```python
def search(pat, txt):
    m = len(pat)
    n = len(txt)

    occurrences = []

    for i in range(n - m + 1):
        j = 0

        while j < m:
            if txt[i + j] != pat[j]:
                break
            j += 1

        if j == m:
            occurrences.append(i)

    return occurrences

txt = "AABAACAADAABAABA"
pat = "AABA"

result = search(pat, txt)

for index in result:
    print("Pattern found at index", index)
```

#Prob-4
Sort a given set of elements using the Insertion sort and determine the time required tosort the elements. Repeat the experiment for different values of n, the number of elements in the list to be sorted and plot a graph of the time taken versus n.  
- Aim :  To Develop a program to implement Insertion sort and determine the time required to sort the elements.  
- Algorithm:  
1. Iterate over the array from the second element to the end (i = 1 to n-1).
2. For each element, compare it to the elements before it (j = i-1 to 0).
3. If the current element is smaller, swap it with the element before it.
4. Repeat step 3 until the current element is no longer smaller than the previous element.
5. Repeat steps 2-4 for all elements in the array.

```python
#Prob_4: write code for insertion sort and plot a graph to visualize the relationship between time taken and input size
import time
import random
import matplotlib.pyplot as plt

# Insertion Sort function
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

# Input sizes
input_sizes = [500, 1000, 2000, 4000, 8000]
time_taken = []

# Measure time
for n in input_sizes:
    arr = [random.randint(1, n) for _ in range(n)]

    start = time.time()
    insertion_sort(arr)
    end = time.time()

    time_taken.append(end - start)

# Plot graph
plt.plot(input_sizes, time_taken, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Time Taken (seconds)")
plt.title("Insertion Sort Time Complexity (O(n))")
plt.grid(True)
plt.show()
```

## Implement Linear Search

### Subtask:
Define a Python function for Linear Search that takes a list and a target element as input and returns the index of the element if found, or -1 otherwise.


**Reasoning**:
I will define the `linear_search` function as described in the instructions, which takes a list and a target element, then searches for the target element using linear search and returns its index or -1 if not found.



```python
def linear_search(arr, target):
    """
    Implements linear search to find a target element in a list.

    Args:
        arr (list): The list to search within.
        target: The element to search for.

    Returns:
        int: The index of the target element if found, otherwise -1.
    """
    for index, element in enumerate(arr):
        if element == target:
            return index
    return -1

print("linear_search function defined successfully.")
```

## Measure Time Complexity

### Subtask:
Create a function to measure the execution time of the Linear Search algorithm. This function should generate lists of varying sizes (n), perform the search, and record the time taken. Ensure to consider both best-case (element at start), worst-case (element at end or not present), and average-case scenarios (element at middle) if possible, or average over random inputs.


**Reasoning**:
The subtask requires defining a function to measure the execution time of the linear search algorithm under different scenarios (best, worst, average case). This involves importing necessary modules like `time` and `random`, generating lists, determining target elements based on the search type, calling the previously defined `linear_search` function, and calculating the elapsed time.



```python
import time
import random

def measure_linear_search_time(n, search_type):
    """
    Measures the execution time of the linear_search algorithm for a given list size (n)
    and search type (best, worst, or average case).

    Args:
        n (int): The size of the list to generate.
        search_type (str): The type of search case ('best', 'worst', 'average').

    Returns:
        float: The time taken to perform the linear search in seconds.
    """
    if n < 0:
        raise ValueError("List size 'n' cannot be negative.")

    # Generate a sorted list of n random integers for consistency, or simply `range(n)`
    # For simplicity and clear case definition, we'll use a range.
    arr = list(range(n))

    target = None
    if search_type == 'best':
        if n == 0: # Handle empty list for best case
            target = None # Not found, but search time should be minimal
        else:
            target = arr[0]  # First element
    elif search_type == 'worst':
        if n == 0: # Handle empty list for worst case
            target = 1 # Not found
        else:
            target = n + 1  # Element not in the list
            # Alternatively, target = arr[-1] to ensure it's found at the very end
            # For a true worst case (not found) target = n + 1 is better.
    elif search_type == 'average':
        if n == 0:
            target = None
        elif n == 1:
            target = arr[0]
        else:
            target = arr[random.randint(0, n - 1)] # A random element from the list
            # For a more consistent 'average', one could pick a fixed middle element
            # target = arr[n // 2]
    else:
        raise ValueError("Invalid search_type. Must be 'best', 'worst', or 'average'.")

    # Measure time
    start_time = time.time()
    linear_search(arr, target)
    end_time = time.time()

    return end_time - start_time

print("measure_linear_search_time function defined successfully.")
```

**Reasoning**:
The next step is to create a function to conduct the experiment, which involves running the `measure_linear_search_time` function for a range of input sizes (`n`) and collecting the results for best, worst, and average cases. This data will then be used for plotting.



```python
def run_experiment_and_collect_data(max_n, step=100, num_runs=5):
    """
    Runs the linear search experiment for various input sizes and collects timing data.

    Args:
        max_n (int): The maximum size of the list to test.
        step (int): The increment for list sizes (e.g., 100 means test for 100, 200, ...).
        num_runs (int): Number of times to run for 'average' case to get a more stable result.

    Returns:
        dict: A dictionary containing lists of N values and corresponding times for best, worst, and average cases.
    """
    n_values = []
    best_case_times = []
    worst_case_times = []
    average_case_times = []

    for n in range(0, max_n + 1, step):
        if n == 0 and step > 0: # Ensure n=0 is included if step allows, otherwise skip
            n_values.append(n)
            # For n=0, search time should be minimal, close to 0
            best_case_times.append(measure_linear_search_time(n, 'best'))
            worst_case_times.append(measure_linear_search_time(n, 'worst'))
            average_case_times.append(measure_linear_search_time(n, 'average'))
            continue
        elif n == 0 and step == 0: # Avoid infinite loop if step is 0 and max_n is 0
            continue
        elif n == 0:
             continue # Skip for now if not explicitly handling for n=0

        n_values.append(n)
        best_case_times.append(measure_linear_search_time(n, 'best'))
        worst_case_times.append(measure_linear_search_time(n, 'worst'))

        # For average case, run multiple times and take the average to smooth out fluctuations
        avg_time_sum = 0
        for _ in range(num_runs):
            avg_time_sum += measure_linear_search_time(n, 'average')
        average_case_times.append(avg_time_sum / num_runs)

    return {
        'n_values': n_values,
        'best_case_times': best_case_times,
        'worst_case_times': worst_case_times,
        'average_case_times': average_case_times
    }

print("run_experiment_and_collect_data function defined successfully.")
```

**Reasoning**:
Now that the functions for linear search, time measurement, and experiment data collection are defined and executed, the next logical step is to use this data to plot a graph, visualizing the relationship between time taken and input size for best, worst, and average cases, as per the main task.



```python
import matplotlib.pyplot as plt

# Run the experiment to get data
max_n = 5000  # Maximum list size to test
step = 100    # Increment for list sizes
num_runs = 5  # Number of runs for average case

print(f"Running experiment for n from 0 to {max_n} with step {step}...")
experiment_data = run_experiment_and_collect_data(max_n, step, num_runs)
print("Experiment data collected successfully.")

n_values = experiment_data['n_values']
best_case_times = experiment_data['best_case_times']
worst_case_times = experiment_data['worst_case_times']
average_case_times = experiment_data['average_case_times']

# Plotting the results
plt.figure(figsize=(12, 8))
plt.plot(n_values, best_case_times, label='Best Case', marker='.', markersize=4)
plt.plot(n_values, worst_case_times, label='Worst Case', marker='.', markersize=4)
plt.plot(n_values, average_case_times, label='Average Case', marker='.', markersize=4)

plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (seconds)')
plt.title('Linear Search Time Complexity vs. Input Size')
plt.legend()
plt.grid(True)
plt.show()

print("Plot generated successfully.")
```
