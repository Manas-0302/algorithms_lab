```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
# func4 counts how many elements in list a also appear in list b.
# It does this by sorting b and using binary search
import time
start_time = time.time()
print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
end_time = time.time()
print("execution time:", end_time - start_time, "seconds")



```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
#this function find the integer square root of n if n is a perfect square
import time
start_time = time.time()
print(func5(256))
end_time = time.time()
print("execution time:", end_time - start_time, "seconds")

```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
# func8 implements the Sort algorithm to sort an array
# It recursively sorts overlapping two-thirds of the array until the whole list is ordered
import time
start_time = time.time()
print(func8([3, 1, 2, 7, 8, 6]))
end_time = time.time()
print("execution time:", end_time - start_time, "seconds")



```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
# func12 rotates the array to the right by k positions
# It builds a new array where each element is shifted k steps forward
import time
start_time = time.time()
print(func12([1, 2, 3, 4, 5], 2))
end_time = time.time()
print("execution time:", end_time - start_time, "seconds")

```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    total = 0
    for x in arr:
        total += x
    return total
print(sum_arr([1, 2, 3, 4, 5]))  # Output: 15
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
arr = [12, 3, 4, 15, 45, 78, 450]
print(sum_arr(arr))
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    minimum = arr[0]
    maximum = arr[0]
    for x in arr:
        if x < minimum:
            minimum = x
        if x > maximum:
            maximum = x
    return minimum, maximum
arr = [12, 3, 4, 15, 45, 78, 450]
print(max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
# Counting_EvenOdd takes an array as input and returns the count of even and odd elements.
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0
    for x in arr:
        if x % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    return even_count, odd_count
arr = [12, 3, 4, 15, 45, 78, 450]
print(Counting_EvenOdd(arr))   # Output: (4, 3)
```

```python
def linear_Search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    low, high = 0, len(arr)
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
print(Binary_Search([1, 3, 5, 7, 9, 15, 16, 14, 45], 15))   # Output: 5
```
